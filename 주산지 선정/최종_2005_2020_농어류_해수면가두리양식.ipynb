```python
농어류_전체_2005 = df_2005[df_2005["품종명"] == "농어류"]
농어류_2005_시군구 = 농어류_전체_2005.groupby("시군구 주소")["양식면적"].sum().reset_index()

농어류_전체_2010 = df_2010[df_2010["품종명"] == "농어"]
농어류_2010_시군구 = 농어류_전체_2010.groupby("시군구 주소")["양식면적"].sum().reset_index()

농어류_전체_2015 = df_2015[df_2015["품종명"] == "농어류"]
농어류_2015_시군구 = 농어류_전체_2015.groupby("시군구 주소")["양식면적"].sum().reset_index()

농어류_전체_2020 = df_2020[df_2020["품종명"] == "농어류"]
농어류_2020_시군구 = 농어류_전체_2020.groupby("시군구 주소")["양식면적"].sum().reset_index()
```

```python
f,ax=plt.subplots(1,4,figsize=(18,4))

# 각 서브플롯에 sns.displot 그리기
sns.histplot(data=농어류_2005_시군구, x="양식면적", bins=15, kde=True, ax=ax[0],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=농어류_2010_시군구, x="양식면적", bins=15, kde=True, ax=ax[1],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=농어류_2015_시군구, x="양식면적", bins=15, kde=True, ax=ax[2],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=농어류_2020_시군구, x="양식면적", bins=15, kde=True, ax=ax[3],color='#b5cae9',line_kws={'lw': 2})
ax[0].lines[0].set_color('#5892e9')
ax[1].lines[0].set_color('#5892e9')
ax[2].lines[0].set_color('#5892e9')
ax[3].lines[0].set_color('#5892e9')

ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[3].set_ylabel('')

ax[0].set_xlabel('2005년 농어류 양식면적', fontsize=15, fontweight='bold')
ax[1].set_xlabel('2010년 농어류 양식면적', fontsize=15, fontweight='bold')
ax[2].set_xlabel('2015년 농어류 양식면적', fontsize=15, fontweight='bold')
ax[3].set_xlabel('2020년 농어류 양식면적', fontsize=15, fontweight='bold')

# 그래프 표시
plt.show()
```

```python
desired_values = ['전라남도 여수시','경상남도 통영시','충청남도 보령시','전라남도 무안군']
desired_values2 = ['전라남도 여수시','경상남도 통영시','경상남도 남해군','전라남도 목포시']
desired_values3 = ['전라남도 여수시','경상남도 거제시','부산광역시 강서구']
desired_values4 = ['전라남도 여수시','경상남도 통영시','충청남도 보령시']
   
```

```python
농어류_05 = df_2005[df_2005['시군구 주소'].isin(desired_values)]
농어류_10 = df_2010[df_2010['시군구 주소'].isin(desired_values2)]
농어류_15 = df_2015[df_2015['시군구 주소'].isin(desired_values3)]
농어류_20 = df_2020[df_2020['시군구 주소'].isin(desired_values4)]
```

```python
농어류_2020= 농어류_20[농어류_20["품종명"] == "농어류"]
농어류_2015= 농어류_15[농어류_15["품종명"] == "농어류"]
농어류_2010= 농어류_10[농어류_10["품종명"] == "농어"]
농어류_2005= 농어류_05[농어류_05["품종명"] == "농어류"]
```

```python
농어류_2020=농어류_2020.reset_index(drop=True)
농어류_20_읍면동 = 농어류_2020.groupby("전체 주소")["양식면적"].sum().reset_index()

농어류_2015=농어류_2015.reset_index(drop=True)
농어류_15_읍면동 = 농어류_2015.groupby("전체 주소")["양식면적"].sum().reset_index()

농어류_2010=농어류_2010.reset_index(drop=True)
농어류_10_읍면동 = 농어류_2010.groupby("전체 주소")["양식면적"].sum().reset_index()

농어류_2005=농어류_2005.reset_index(drop=True)
농어류_05_읍면동 = 농어류_2005.groupby("전체 주소")["양식면적"].sum().reset_index()
```

```python
농어류_05_읍면동.sort_values(by='양식면적', ascending=False)
```

```python
f,ax=plt.subplots(1,4,figsize=(18,4))

# 각 서브플롯에 sns.displot 그리기
sns.histplot(data=농어류_05_읍면동, x="양식면적", bins=50, kde=True, ax=ax[0],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=농어류_10_읍면동, x="양식면적", bins=50, kde=True, ax=ax[1],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=농어류_15_읍면동, x="양식면적", bins=50, kde=True, ax=ax[2],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=농어류_20_읍면동, x="양식면적", bins=50, kde=True, ax=ax[3],color='#d0a8f2',line_kws={'lw': 2})
ax[0].lines[0].set_color('purple')
ax[1].lines[0].set_color('purple')
ax[2].lines[0].set_color('purple')
ax[3].lines[0].set_color('purple')
ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[3].set_ylabel('')

ax[0].set_xlabel('2005년 농어 주산지 양식면적', fontsize=15, fontweight='bold')
ax[1].set_xlabel('2010년 농어 주산지 양식면적', fontsize=15, fontweight='bold')
ax[2].set_xlabel('2015년 농어 주산지 양식면적', fontsize=15, fontweight='bold')
ax[3].set_xlabel('2015년 농어 주산지 양식면적', fontsize=15, fontweight='bold')

# 그래프 표시
plt.show()
```
