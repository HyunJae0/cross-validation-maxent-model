```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df_2020=pd.read_csv("2020년해상가두리양식.csv", encoding='cp949')
df_2015=pd.read_csv("2015년해상가두리양식.csv", encoding='cp949')
df_2010=pd.read_csv("2010년해상가두리양식.csv", encoding='cp949')
df_2005=pd.read_csv("2005년해상가두리양식2.csv", encoding='cp949')
```

```python
df_2005
```

```python
df_2005['시군구 주소'] = df_2005['전체 주소'].apply(lambda x: ' '.join(x.split(' ')[:2]))
```

```python
df_2005
```

```python
df_2005['품종명'].unique()
```

```python
참돔_전체_2005 = df_2005[df_2005["품종명"] == "참돔"]
참돔_2005_시군구 = 참돔_전체_2005.groupby("시군구 주소")["양식면적"].sum().reset_index()

참돔_전체_2010 = df_2010[df_2010["품종명"] == "참돔"]
참돔_2010_시군구 = 참돔_전체_2010.groupby("시군구 주소")["양식면적"].sum().reset_index()

참돔_전체_2015 = df_2015[df_2015["품종명"] == "참돔"]
참돔_2015_시군구 = 참돔_전체_2015.groupby("시군구 주소")["양식면적"].sum().reset_index()

참돔_전체_2020 = df_2020[df_2020["품종명"] == "참돔"]
참돔_2020_시군구 = 참돔_전체_2020.groupby("시군구 주소")["양식면적"].sum().reset_index()
```

```python
f,ax=plt.subplots(1,4,figsize=(20,4))

# 각 서브플롯에 sns.displot 그리기
sns.histplot(data=참돔_2005_시군구, x="양식면적", bins=15, kde=True, ax=ax[0],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=참돔_2010_시군구, x="양식면적", bins=15, kde=True, ax=ax[1],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=참돔_2015_시군구, x="양식면적", bins=15, kde=True, ax=ax[2],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=참돔_2020_시군구, x="양식면적", bins=15, kde=True, ax=ax[3],color='#b5cae9',line_kws={'lw': 2})
ax[0].lines[0].set_color('#5892e9')
ax[1].lines[0].set_color('#5892e9')
ax[2].lines[0].set_color('#5892e9')
ax[3].lines[0].set_color('#5892e9')

ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[3].set_ylabel('')

ax[0].set_xlabel('2005년 참돔 양식면적', fontsize=15, fontweight='bold')
ax[1].set_xlabel('2010년 참돔 양식면적', fontsize=15, fontweight='bold')
ax[2].set_xlabel('2015년 참돔 양식면적', fontsize=15, fontweight='bold')
ax[3].set_xlabel('2020년 참돔 양식면적', fontsize=15, fontweight='bold')

# 그래프 표시
plt.show()
```

```python
desired_values = ['경상남도 통영시','전라남도 여수시','경상남도 남해군','경상남도 거제시']
desired_values2 = ['경상남도 통영시','전라남도 여수시','경상남도 남해군','경상남도 거제시','전라남도 완도군']
```

```python
참돔_05 = df_2005[df_2005['시군구 주소'].isin(desired_values)]
참돔_10 = df_2010[df_2010['시군구 주소'].isin(desired_values2)]
참돔_15 = df_2015[df_2015['시군구 주소'].isin(desired_values)]
참돔_20 = df_2020[df_2020['시군구 주소'].isin(desired_values)]
```

```python
참돔_2020= 참돔_20[참돔_20["품종명"] == "참돔"]
참돔_2015= 참돔_15[참돔_15["품종명"] == "참돔"]
참돔_2010= 참돔_10[참돔_10["품종명"] == "참돔"]
참돔_2005= 참돔_05[참돔_05["품종명"] == "참돔"]
```

```python
참돔_2020=참돔_2020.reset_index(drop=True)
참돔_20_읍면동 = 참돔_2020.groupby("전체 주소")["양식면적"].sum().reset_index()

참돔_2015=참돔_2015.reset_index(drop=True)
참돔_15_읍면동 = 참돔_2015.groupby("전체 주소")["양식면적"].sum().reset_index()

참돔_2010=참돔_2010.reset_index(drop=True)
참돔_10_읍면동 = 참돔_2010.groupby("전체 주소")["양식면적"].sum().reset_index()

참돔_2005=참돔_2005.reset_index(drop=True)
참돔_05_읍면동 = 참돔_2005.groupby("전체 주소")["양식면적"].sum().reset_index()
```

```python
참돔_05_읍면동.sort_values(by='양식면적', ascending=False)
```

```python
f,ax=plt.subplots(1,4,figsize=(18,4))

# 각 서브플롯에 sns.displot 그리기
sns.histplot(data=참돔_05_읍면동, x="양식면적", bins=50, kde=True, ax=ax[0],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=참돔_10_읍면동, x="양식면적", bins=50, kde=True, ax=ax[1],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=참돔_15_읍면동, x="양식면적", bins=50, kde=True, ax=ax[2],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=참돔_20_읍면동, x="양식면적", bins=50, kde=True, ax=ax[3],color='#d0a8f2',line_kws={'lw': 2})
ax[0].lines[0].set_color('purple')
ax[1].lines[0].set_color('purple')
ax[2].lines[0].set_color('purple')
ax[3].lines[0].set_color('purple')
ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[3].set_ylabel('')

ax[0].set_xlabel('2005년 전복 주산지 양식면적', fontsize=15, fontweight='bold')
ax[1].set_xlabel('2010년 전복 주산지 양식면적', fontsize=15, fontweight='bold')
ax[2].set_xlabel('2015년 전복 주산지 양식면적', fontsize=15, fontweight='bold')
ax[3].set_xlabel('2015년 전복 주산지 양식면적', fontsize=15, fontweight='bold')

# 그래프 표시
plt.show()
```
