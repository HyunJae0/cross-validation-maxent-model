```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df_2020=pd.read_csv("2020년해상가두리양식.csv", encoding='cp949')
df_2015=pd.read_csv("2015년해상가두리양식.csv", encoding='cp949')
df_2010=pd.read_csv("2010년해상가두리양식.csv", encoding='cp949')
df_2005=pd.read_csv("2005년해상가두리양식2.csv", encoding='cp949')
```

```python
df_2005['시군구 주소'] = df_2005['전체 주소'].apply(lambda x: ' '.join(x.split(' ')[:2]))
```

```python
df_2005
```

```python
숭어류_전체_2005 = df_2005[df_2005["품종명"] == "숭어류"]
숭어류_2005_시군구 = 숭어류_전체_2005.groupby("시군구 주소")["양식면적"].sum().reset_index()

숭어류_전체_2010 = df_2010[df_2010["품종명"] == "숭어류"]
숭어류_2010_시군구 = 숭어류_전체_2010.groupby("시군구 주소")["양식면적"].sum().reset_index()

숭어류_전체_2015 = df_2015[df_2015["품종명"] == "숭어류"]
숭어류_2015_시군구 = 숭어류_전체_2015.groupby("시군구 주소")["양식면적"].sum().reset_index()

숭어류_전체_2020 = df_2020[df_2020["품종명"] == "숭어류"]
숭어류_2020_시군구 = 숭어류_전체_2020.groupby("시군구 주소")["양식면적"].sum().reset_index()
```

```python
f,ax=plt.subplots(1,4,figsize=(18,4))

# 각 서브플롯에 sns.displot 그리기
sns.histplot(data=숭어류_2005_시군구, x="양식면적", bins=15, kde=True, ax=ax[0],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=숭어류_2010_시군구, x="양식면적", bins=15, kde=True, ax=ax[1],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=숭어류_2015_시군구, x="양식면적", bins=15, kde=True, ax=ax[2],color='#b5cae9',line_kws={'lw': 2})
sns.histplot(data=숭어류_2020_시군구, x="양식면적", bins=15, kde=True, ax=ax[3],color='#b5cae9',line_kws={'lw': 2})
ax[0].lines[0].set_color('#5892e9')
ax[1].lines[0].set_color('#5892e9')
ax[2].lines[0].set_color('#5892e9')
ax[3].lines[0].set_color('#5892e9')

ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[3].set_ylabel('')

ax[0].set_xlabel('2005년 숭어류 양식면적', fontsize=15, fontweight='bold')
ax[1].set_xlabel('2010년 숭어류 양식면적', fontsize=15, fontweight='bold')
ax[2].set_xlabel('2015년 숭어류 양식면적', fontsize=15, fontweight='bold')
ax[3].set_xlabel('2020년 숭어류 양식면적', fontsize=15, fontweight='bold')

# 그래프 표시
plt.show()
```

```python
desired_values = ['경상남도 하동군','전라북도 고창군','경상남도 남해군','전라남도 영광군','전라남도 무안군']
desired_values2 = ['경상남도 하동군', '충청남도 태안군','경상남도 진주시','충청남도 서산시']
desired_values3 = ['경상남도 하동군', '충청남도 태안군','충청남도 당진시','충청남도 서산시']
desired_values4 = ['전라남도 여수시','충청남도 서산시','경상남도 하동군']
```

```python
숭어류_05 = df_2005[df_2005['시군구 주소'].isin(desired_values)]
숭어류_10 = df_2010[df_2010['시군구 주소'].isin(desired_values2)]
숭어류_15 = df_2015[df_2015['시군구 주소'].isin(desired_values3)]
숭어류_20 = df_2020[df_2020['시군구 주소'].isin(desired_values4)]
```

```python
숭어류_2020= 숭어류_20[숭어류_20["품종명"] == "숭어류"]
숭어류_2015= 숭어류_15[숭어류_15["품종명"] == "숭어류"]
숭어류_2010= 숭어류_10[숭어류_10["품종명"] == "숭어류"]
숭어류_2005= 숭어류_05[숭어류_05["품종명"] == "숭어류"]
```

```python
숭어류_2020=숭어류_2020.reset_index(drop=True)
숭어류_20_읍면동 = 숭어류_2020.groupby("전체 주소")["양식면적"].sum().reset_index()

숭어류_2015=숭어류_2015.reset_index(drop=True)
숭어류_15_읍면동 = 숭어류_2015.groupby("전체 주소")["양식면적"].sum().reset_index()

숭어류_2010=숭어류_2010.reset_index(drop=True)
숭어류_10_읍면동 = 숭어류_2010.groupby("전체 주소")["양식면적"].sum().reset_index()

숭어류_2005=숭어류_2005.reset_index(drop=True)
숭어류_05_읍면동 = 숭어류_2005.groupby("전체 주소")["양식면적"].sum().reset_index()
```
