```python
df_2020=pd.read_csv("2020년해상가두리양식.csv", encoding='cp949')
df_2015=pd.read_csv("2015년해상가두리양식.csv", encoding='cp949')
df_2010=pd.read_csv("2010년해상가두리양식.csv", encoding='cp949')
df_2005=pd.read_csv("2005년해상가두리양식2.csv", encoding='cp949')
```

```python
감성돔_전체_2005 = df_2005[df_2005["품종명"] == "감성돔"]
감성돔_2005_시군구 = 감성돔_전체_2005.groupby("시군구 주소")["양식면적"].sum().reset_index()

감성돔_전체_2010 = df_2010[df_2010["품종명"] == "감성돔"]
감성돔_2010_시군구 = 감성돔_전체_2010.groupby("시군구 주소")["양식면적"].sum().reset_index()

감성돔_전체_2015 = df_2015[df_2015["품종명"] == "감성돔"]
감성돔_2015_시군구 = 감성돔_전체_2015.groupby("시군구 주소")["양식면적"].sum().reset_index()

감성돔_전체_2020 = df_2020[df_2020["품종명"] == "감성돔"]
감성돔_2020_시군구 = 감성돔_전체_2020.groupby("시군구 주소")["양식면적"].sum().reset_index()
```

```python
desired_values1 = ['경상남도 통영시','전라남도 여수시','전라남도 완도군']
desired_values2 = ['경상남도 통영시','전라남도 여수시','경상남도 남해군']
desired_values3 = ['경상남도 통영시','전라남도 여수시','전라북도 부안군']
```

```python
감성돔_05 = df_2005[df_2005['시군구 주소'].isin(desired_values1)]
감성돔_10 = df_2010[df_2010['시군구 주소'].isin(desired_values2)]
감성돔_15 = df_2015[df_2015['시군구 주소'].isin(desired_values3)]
감성돔_20 = df_2020[df_2020['시군구 주소'].isin(desired_values2)]
```

```python
감성돔_15['시군구 주소'].unique()
```

```python
감성돔_2020= 감성돔_20[감성돔_20["품종명"] == "감성돔"]
감성돔_2015= 감성돔_15[감성돔_15["품종명"] == "감성돔"]
감성돔_2010= 감성돔_10[감성돔_10["품종명"] == "감성돔"]
감성돔_2005= 감성돔_05[감성돔_05["품종명"] == "감성돔"]
```

```python
감성돔_2020=감성돔_2020.reset_index(drop=True)
감성돔_20_읍면동 = 감성돔_2020.groupby("전체 주소")["양식면적"].sum().reset_index()

감성돔_2015=감성돔_2015.reset_index(drop=True)
감성돔_15_읍면동 = 감성돔_2015.groupby("전체 주소")["양식면적"].sum().reset_index()

감성돔_2010=감성돔_2010.reset_index(drop=True)
감성돔_10_읍면동 = 감성돔_2010.groupby("전체 주소")["양식면적"].sum().reset_index()

감성돔_2005=감성돔_2005.reset_index(drop=True)
감성돔_05_읍면동 = 감성돔_2005.groupby("전체 주소")["양식면적"].sum().reset_index()
```

```python
감성돔_05_읍면동.sort_values(by='양식면적', ascending=False)
```

```python
f,ax=plt.subplots(1,4,figsize=(18,4))

# 각 서브플롯에 sns.displot 그리기
sns.histplot(data=감성돔_05_읍면동, x="양식면적", bins=50, kde=True, ax=ax[0],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=감성돔_10_읍면동, x="양식면적", bins=50, kde=True, ax=ax[1],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=감성돔_15_읍면동, x="양식면적", bins=50, kde=True, ax=ax[2],color='#d0a8f2',line_kws={'lw': 2})
sns.histplot(data=감성돔_20_읍면동, x="양식면적", bins=50, kde=True, ax=ax[3],color='#d0a8f2',line_kws={'lw': 2})
ax[0].lines[0].set_color('purple')
ax[1].lines[0].set_color('purple')
ax[2].lines[0].set_color('purple')
ax[3].lines[0].set_color('purple')
ax[1].set_ylabel('')
ax[2].set_ylabel('')
ax[3].set_ylabel('')

ax[0].set_xlabel('2005년 전복 주산지 양식면적', fontsize=15, fontweight='bold')
ax[1].set_xlabel('2010년 전복 주산지 양식면적', fontsize=15, fontweight='bold')
ax[2].set_xlabel('2015년 전복 주산지 양식면적', fontsize=15, fontweight='bold')
ax[3].set_xlabel('2015년 전복 주산지 양식면적', fontsize=15, fontweight='bold')

# 그래프 표시
plt.show()
```
